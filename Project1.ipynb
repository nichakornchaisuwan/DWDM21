#### เปลี่ยนชื่อคอลัมน์ data_3New

````python
data_3New.rename(columns={'2555':'2555income'},inplace=True)
data_3New.rename(columns={'2556':'2556income'},inplace=True)
data_3New.rename(columns={'2557':'2557income'},inplace=True)
data_3New.rename(columns={'2558':'2558income'},inplace=True)
data_3New.rename(columns={'2559':'2559income'},inplace=True)
data_3New.rename(columns={'2560':'2560income'},inplace=True)
data_3New.rename(columns={'2561':'2561income'},inplace=True)
````
